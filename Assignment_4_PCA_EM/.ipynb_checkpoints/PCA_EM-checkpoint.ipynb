# PCA

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from fpdf import FPDF
import umap.umap_ as umap



# Create a PDF class
pdf = FPDF()
pdf.add_page()
pdf.set_font(family="Times",style="b", size=18)


# Load PCA data
pca_data = np.loadtxt("pca_data.txt")

# Step 1: Standardize data
mean = np.mean(pca_data, axis=0)
std = np.std(pca_data, axis=0)
standardized_data = (pca_data - mean) / std

# Step 2: Compute the covariance matrix
cov_matrix = np.cov(standardized_data, rowvar=False)

# Step 3: Eigenvalue decomposition
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# Step 4: Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Step 5: Project data onto the top two eigenvectors
top_2_eigenvectors = eigenvectors[:, :2]
projected_data = np.dot(standardized_data, top_2_eigenvectors)

# Step 6: Create 2D scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(projected_data[:, 0], projected_data[:, 1], alpha=0.5, s=30)
plt.title("2D Scatter Plot of PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.savefig(f"2D_Scatter_PCA_plot.png")
plt.show()


# Add a page to the PDF
pdf.add_page()
pdf.image(f"2D_Scatter_PCA_plot.png", x=10, y=10, w=180, h=0)

# Step 7: Create UMAP and t-SNE plots
# UMAP
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42)
umap_embedding = reducer.fit_transform(standardized_data)
plt.figure(figsize=(8, 6))
plt.scatter(umap_embedding[:, 0], umap_embedding[:, 1], alpha=0.5, s=30)
plt.title("UMAP Plot")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.grid(True)
plt.savefig(f"UMAP_plot.png")
plt.show()


# Add a page to the PDF
pdf.add_page()
pdf.image(f"UMAP_plot.png", x=10, y=10, w=180, h=0)

# t-SNE
tsne = TSNE(n_components=2, random_state=42)
tsne_embedding = tsne.fit_transform(standardized_data)
plt.figure(figsize=(8, 6))
plt.scatter(tsne_embedding[:, 0], tsne_embedding[:, 1], alpha=0.5, s=30)
plt.title("t-SNE Plot")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.grid(True)
plt.savefig(f"t-SNE_plot.png")
plt.show()

# Add a page to the PDF
pdf.add_page()
pdf.image(f"t-SNE_plot.png", x=10, y=10, w=180, h=0)
```

# EM

```python
from scipy.stats import poisson

# Load EM data
em_data = np.loadtxt("em_data.txt")

# Step 1: Initialize parameters
np.random.seed(42)
mu_1, mu_2 = np.random.uniform(1, 5, 2)  # Means
pi_1 = 0.5  # Initial proportion of families with planning advice
pi_2 = 1 - pi_1

tolerance = 1e-4
max_iter = 100
prev_mu_1, prev_mu_2 = 0, 0

# Step 2: EM Algorithm
for iteration in range(max_iter):
    # E-step: Calculate responsibilities
    r1 = pi_1 * poisson.pmf(em_data, mu_1)
    r2 = pi_2 * poisson.pmf(em_data, mu_2)
    total_r = r1 + r2
    gamma_1 = r1 / total_r
    gamma_2 = r2 / total_r

    # M-step: Update parameters
    mu_1 = np.sum(gamma_1 * em_data) / np.sum(gamma_1)
    mu_2 = np.sum(gamma_2 * em_data) / np.sum(gamma_2)
    pi_1 = np.mean(gamma_1)
    pi_2 = np.mean(gamma_2)

    # Check convergence
    if abs(mu_1 - prev_mu_1) < tolerance and abs(mu_2 - prev_mu_2) < tolerance:
        break
    prev_mu_1, prev_mu_2 = mu_1, mu_2

# Step 3: Report results
print(f"The Estimation converged after {iteration} iterations.")
print("Estimated Parameters for EM:")
print(f"Mean number of children (Family Planning): {mu_1:.2f}")
print(f"Mean number of children (No Family Planning): {mu_2:.2f}")
print(f"Proportion of families with Family Planning: {pi_1:.2f}")
print(f"Proportion of families without Family Planning: {pi_2:.2f}")


em_report = []  
em_report.append(
    f"The Estimation converged after {iteration} iterations.\n"
    f"Estimated Parameters for EM:\n"
    f"Mean number of children (Family Planning): {mu_1:.2f}\n"
    f"Mean number of children (No Family Planning): {mu_2:.2f}\n"
    f"Proportion of families with Family Planning: {pi_1:.2f}\n"
    f"Proportion of families without Family Planning: {pi_2:.2f}"
)

# Convert the list into a string
em_report_str = "\n".join(em_report)

# Add it to the PDF
pdf.add_page()
pdf.set_font(family="Times", style="b", size=18)
pdf.cell(0, 10, "", ln=True)  # Blank line
pdf.cell(0, 10, "EM Reports:", ln=True)  # Heading for the test results
pdf.multi_cell(0, 10, em_report_str)  # Add test results to the PDF


pdf.output("PCA_EM_report.pdf")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Function to load data
def load_data(file_path):
    return np.loadtxt(file_path)

# Function to initialize parameters
def initialize_parameters(data):
    np.random.seed(42)
    lambda1 = np.random.rand() * np.mean(data)  # Initial Poisson mean for family planning
    lambda2 = np.random.rand() * np.mean(data)  # Initial Poisson mean for no family planning
    pi = 0.5  # Initial proportion of families with family planning
    return lambda1, lambda2, pi

# Function to perform the E-step (responsibilities)
def e_step(data, lambda1, lambda2, pi):
    p1 = pi * poisson.pmf(data, lambda1)
    p2 = (1 - pi) * poisson.pmf(data, lambda2)
    responsibilities = p1 / (p1 + p2)
    return responsibilities, p1, p2

# Function to perform the M-step (parameter updates)
def m_step(data, responsibilities):
    lambda1_new = np.sum(responsibilities * data) / np.sum(responsibilities)
    lambda2_new = np.sum((1 - responsibilities) * data) / np.sum(1 - responsibilities)
    pi_new = np.mean(responsibilities)
    return lambda1_new, lambda2_new, pi_new

# Function to compute log-likelihood
def compute_log_likelihood(p1, p2):
    return np.sum(np.log(p1 + p2))

# Function to plot log-likelihood convergence
def plot_log_likelihood(log_likelihoods):
    plt.figure(figsize=(8, 6))
    plt.plot(log_likelihoods, marker='o')
    plt.title("Log-Likelihood Convergence in EM Algorithm")
    plt.xlabel("Iteration")
    plt.ylabel("Log-Likelihood")
    plt.grid()
    plt.show()

# Function to plot estimated Poisson distributions
def plot_poisson_distributions(data, lambda1, lambda2):
    x = np.arange(0, np.max(data) + 1)
    poisson1 = poisson.pmf(x, lambda1)
    poisson2 = poisson.pmf(x, lambda2)

    plt.figure(figsize=(8, 6))
    plt.bar(x, poisson1, alpha=0.6, label=f"Family Planning (={lambda1:.2f})", color="blue")
    plt.bar(x, poisson2, alpha=0.6, label=f"No Family Planning (={lambda2:.2f})", color="orange")
    plt.title("Estimated Poisson Distributions")
    plt.xlabel("Number of Children")
    plt.ylabel("Probability")
    plt.legend()
    plt.grid()
    plt.show()

# Main function to run the EM algorithm
def run_em_algorithm(file_path):
    # Load data
    data = load_data(file_path)
    
    # Initialize parameters
    lambda1, lambda2, pi = initialize_parameters(data)
    
    tolerance = 1e-6
    max_iterations = 100
    log_likelihoods = []
    
    # EM Algorithm
    for iteration in range(max_iterations):
        # E-step: Compute responsibilities
        responsibilities, p1, p2 = e_step(data, lambda1, lambda2, pi)
        
        # M-step: Update parameters
        lambda1_new, lambda2_new, pi_new = m_step(data, responsibilities)
        
        # Compute log-likelihood
        log_likelihood = compute_log_likelihood(p1, p2)
        log_likelihoods.append(log_likelihood)
        
        # Check for convergence
        if iteration > 0 and abs(log_likelihood - log_likelihoods[-1]) < tolerance:
            break
        
        # Update parameters
        lambda1, lambda2, pi = lambda1_new, lambda2_new, pi_new
    
    # Plot Log-Likelihood Convergence
    plot_log_likelihood(log_likelihoods)
    
    # Plot Estimated Poisson Distributions
    plot_poisson_distributions(data, lambda1, lambda2)
    
    # Print results
    print(f"Lambda for Family Planning (1): {lambda1:.2f}")
    print(f"Lambda for No Family Planning (2): {lambda2:.2f}")
    print(f"Proportion of families with Family Planning: {pi:.2f}")
    print(f"Proportion of families without Family Planning: {1 - pi:.2f}")

# Call the main function with the path to your EM data file
if __name__ == "__main__":
    run_em_algorithm("em_data.txt")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Step 1: Load the data
def load_data(file_path):
    return np.loadtxt(file_path)

# Step 2: Initialize parameters
def initialize_parameters(data):
    np.random.seed(42)
    lambda1, lambda2 = np.random.rand(2) * np.mean(data)
    pi = 0.5  # Initial proportion
    return lambda1, lambda2, pi

# Step 3: Perform the E-step
def e_step(data, lambda1, lambda2, pi):
    p1 = pi * poisson.pmf(data, lambda1)
    p2 = (1 - pi) * poisson.pmf(data, lambda2)
    responsibilities = p1 / (p1 + p2)
    return responsibilities

# Step 4: Perform the M-step
def m_step(data, responsibilities):
    lambda1 = np.sum(responsibilities * data) / np.sum(responsibilities)
    lambda2 = np.sum((1 - responsibilities) * data) / np.sum(1 - responsibilities)
    pi = np.mean(responsibilities)
    return lambda1, lambda2, pi

# Step 5: Run the EM algorithm
def em_algorithm(data, tolerance=1e-6, max_iterations=100):
    lambda1, lambda2, pi = initialize_parameters(data)
    for iteration in range(max_iterations):
        # E-step
        responsibilities = e_step(data, lambda1, lambda2, pi)

        # M-step
        lambda1_new, lambda2_new, pi_new = m_step(data, responsibilities)

        # Check for convergence
        if abs(lambda1 - lambda1_new) < tolerance and abs(lambda2 - lambda2_new) < tolerance:
            break

        lambda1, lambda2, pi = lambda1_new, lambda2_new, pi_new

    return lambda1, lambda2, pi

# Step 6: Plot the data and fitted mixture model
def plot_mixture_model(data, lambda1, lambda2, pi):
    plt.figure(figsize=(10, 6))
    bins = np.arange(0, np.max(data) + 2) - 0.5
    plt.hist(data, bins=bins, density=True, alpha=0.6, color="gray", label="Observed Data")

    # Generate Poisson distributions
    x = np.arange(0, np.max(data) + 1)
    poisson1 = pi * poisson.pmf(x, lambda1)
    poisson2 = (1 - pi) * poisson.pmf(x, lambda2)
    mixture = poisson1 + poisson2

    # Plot the distributions
    plt.plot(x, poisson1, 'o-', color='blue', label=f"Family Planning (={lambda1:.2f})")
    plt.plot(x, poisson2, 'o-', color='orange', label=f"No Family Planning (={lambda2:.2f})")
    plt.plot(x, mixture, 'o-', color='green', label="Mixture Model")

    # Customize the plot
    plt.title("Fitted Poisson Mixture Model")
    plt.xlabel("Number of Children")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()

# Main function to tie everything together
def main():
    # Load the dataset
    data = load_data("em_data.txt")

    # Run the EM algorithm
    lambda1, lambda2, pi = em_algorithm(data)

    # Print results
    print(f"Estimated 1 (Family Planning): {lambda1:.2f}")
    print(f"Estimated 2 (No Family Planning): {lambda2:.2f}")
    print(f"Proportion of families with family planning: {pi:.2f}")
    print(f"Proportion of families without family planning: {1 - pi:.2f}")

    # Plot the results
    plot_mixture_model(data, lambda1, lambda2, pi)

# Run the main function
if __name__ == "__main__":
    main()
```
